```python
#import pandas libraries :
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset
epl_df = pd.read_csv('C:\\Users\\JANESAR\\Desktop\\EPL_20_21.csv')
epl_df.head()
```

```python
# Information about Dataset
epl_df.info()
```

```python
# Summary statistics about datasets : It only works on numeric columns :
epl_df.describe()
```

# Check null values :

```python

epl_df.isna().sum()
```

Finding Match per Minutes, and Goals per Matches : 

```python
epl_df["MinsperMatch"] = (epl_df["Mins"] / epl_df["Matches"]).astype(int)
epl_df["GoalsperMatch"] = (epl_df["Goals"]) / epl_df["Matches"].astype(float)
epl_df.head()
```

Calculate the total Goals in EPL_2021 seasons :

```python
Total_Goals = epl_df["Goals"].sum()
Total_Goals
```

 Calculate total penalty Goals :

```python
Total_penalty_Goals = epl_df["Penalty_Goals"].sum()
Total_penalty_Goals
```

Calculate Total penalty attempts :

```python
Total_penalty_Attempts = epl_df["Penalty_Attempted"].sum()
Total_penalty_Attempts
```

Create pie charts for penalty missed and scored :

```python
plt.figure(figsize=(8, 3))
pl_not_scored = epl_df["Penalty_Attempted"].sum()-Total_penalty_Goals
data = [pl_not_scored, Total_penalty_Goals]
labels = ["Penalties missed", "Penalty Scored"]
color = sns.color_palette("Set1")
plt.pie(data, labels = labels, colors=color, autopct="%.0f%%")
plt.show()
```

Find the most of the players from which countries :?

Which clubs with the maximum players in the squad : ?

```python
epl_df["Club"].value_counts().nlargest(5).plot(kind="bar",figsize=(6, 4), color=sns.color_palette("viridis"))
```

```python
epl_df["Club"].value_counts().nsmallest(5).plot(kind="bar", figsize=(6, 3), color=sns.color_palette("magma"))
```

Player based on age group : ?

```python
age_under20 = epl_df[epl_df["Age"]<=20]
age_under20_25 = epl_df[(epl_df["Age"]>20) & (epl_df["Age"]<=25)]
age_under25_30 = epl_df[(epl_df["Age"]>25) & (epl_df["Age"]<=30)]
age_Above30 = epl_df[epl_df["Age"]>30]
```

```python
x= np.array([age_under20["Name"].count(),age_under20_25["Name"].count(),age_under25_30["Name"].count(),age_Above30["Name"].count()])
labels = ["<=20", ">20 & <=25", ">25& <=30", ">30"]
plt.title("Total players with Age groups", fontsize=20)
plt.pie(x, labels= labels, autopct="%.0f%%" )
plt.show()
```

Total under age 20 of each club :?

Average Age player in each clubs :? 

```python
plt.figure(figsize=(8, 4))
plt.title("Average age players in each clubs, fontsize=20")
sns.boxplot(x="Club", y="Age", data=epl_df)
plt.xticks(rotation=90)
```

```python
number_of_players = epl_df.groupby("Club").size()
data = (epl_df.groupby("Club")["Age"].sum())/number_of_players
data.sort_values(ascending=False)
```

Total Assists from each club : ?

```python
Assits_by_clubs = pd.DataFrame(epl_df.groupby("Club", as_index=False)["Assists"].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x="Club", y="Assists", data =Assits_by_clubs.sort_values(by="Assists"), palette="Set2")
# data = (Assists_by_clubs.sort_values(by="Assists"), palette="Set2")
ax.set_xlabel("Club", fontsize=30)
ax.set_ylabel("Assists", fontsize=20)
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"]=(20, 7)
plt.title("Plot od clubs VS total Assists", fontsize=20)
```

Top 10 Assists :

```python
top_10_Assists = epl_df[["Name", "Club", "Assists", "Matches"]].nlargest(n=10, columns="Assists")
# top_10_Assists.head()
top_10_Assists
```

Totals Goals from each Clubs :

```python
Goals_by_clubs = pd.DataFrame(epl_df.groupby("Club", as_index=False)["Goals"].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x="Club", y="Goals", data=Goals_by_clubs.sort_values(by="Goals"), palette="rocket")
ax.set_xlabel("Club", fontsize=30)
ax.set_ylabel("Goals", fontsize=20)
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"]=(10, 5)
plt.title("Plot of Clubs VS total Goals", fontsize=20)
```

Most Goals by players :

```python
most_goals_by_player = epl_df[["Name", "Club", "Goals", "Matches", "Position"]].nlargest(n=10, columns="Goals")
most_goals_by_player
```

Top 10 Goals by players in Permatch :

Plot pie chart :- Goals with assists and without assists. 

```python
plt.figure(figsize=(5, 4))
assists = epl_df["Assists"].sum()
data = [Total_Goals - assists, assists]
labels = ["Goals without assists", "goals with assists"]
color = sns.color_palette("Set1")
plt.title("Goals assists VS without Assists", fontsize=15)
plt.pie(data, labels=labels, colors=color, autopct="%.0f%%")
plt.show()
```

Top most 10 players received Yellow card in EPL last season :

```python
yellowcard_in_epl = epl_df.sort_values(by="Yellow_Cards", ascending=False)[: 10]
plt.figure(figsize=(10, 5))
plt.title("Players with max yellow cards : ", fontsize=15)
Rectangle =sns.barplot(x=yellowcard_in_epl["Name"], y=yellowcard_in_epl["Yellow_Cards"], label="Players", color="Yellow")
plt.ylabel("Number of yellow cards")
Rectangle.set_xticklabels(Rectangle.get_xticklabels(), rotation=75)
Rectangle
```
